```python
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

labels = ["Python","C++","Java","JS","C","C#"]
index = np.arange(len(labels))
ratings = [5.16,5.73,14.99,3.17,11.86,4.45]
change = [1.12,0.3,-1.69,0.29,3.41,-0.45]
```

```python
plt.bar(index,ratings)
plt.xticks(index,labels)
plt.ylabel("Change")
plt.title("Programming rating")
plt.show()
```

```python
plt.barh(index,change)
plt.xticks(index,labels)
plt.ylabel("Change")
plt.title("Programming rating")
plt.show()
```

```python
x = np.linspace(0,10,50)
sinus = np.sin(x)
sinhs = np.sinh(x)
```

```python
fig, ax = plt.subplots()
ax.plot(x,sinus,'r-o')
ax2 = ax.twinx()
ax2.plot(x,sinhs,'g--')
plt.show()
```

```python
x = np.linspace(0, 10, 50)
sinus = np.sin(x)
cosinus = np.cos(x)
tangent = np.tan(x)
cotangent = np.tanh(x)

fig, ax = plt.subplots(2,2)
ax[0][0].plot(x, sinus, "^-b", label="sin(x)")
ax[0][1].plot(x, cosinus, "o--m", label="cos(x)")
ax[1][0].plot(x, tangent, "^-b", label="tan(x)")
ax[1][1].plot(x, cotangent, "o--m", label="tanh(x)")

#plt.subplot(4, 1, 1)
#plt.plot(x, sinus, "^-b", label="sin(x)")
#plt.subplot(4, 1, 2)
#plt.plot(x, cosinus, "o--m", label="cos(x)")
#plt.subplot(4, 2, 1)
#plt.plot(x, tangent, "^-b", label="tan(x)")
#plt.subplot(4, 2, 2)
#plt.plot(x, cotangent, "o--m", label="sin(x)")
plt.show()
```
